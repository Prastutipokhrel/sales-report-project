## Step 1: Load the Raw Dataset


```python
data = pd.read_csv("../data/raw_sales_data.csv", encoding='latin1')
data.head()
```

## Step 2 : cleaning the data


```python
#Checking for missing values
data.isnull().sum()
```

```python
data.drop(columns=['ADDRESSLINE2', 'TERRITORY'], inplace=True)
data = data.dropna(subset=['POSTALCODE'])
data['STATE'] = data['STATE'].fillna('Unknown')
```

```python
data['ORDERDATE'] = pd.to_datetime(data['ORDERDATE'])
```

```python
data['Total Sale'] = data['QUANTITYORDERED'] * data['PRICEEACH']
```

## Step 3: Save the Cleaned Dataset


```python
data.to_csv('../data/cleaned_sales_data.csv', index=False)
print("Cleaned file saved!")
```
